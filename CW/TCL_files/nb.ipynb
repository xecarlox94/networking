```python
import tracefile


df2s = tracefile.get_df("top2_s.tr")
df2m = tracefile.get_df("top2_m.tr")
df2l = tracefile.get_df("top2_l.tr")
df1s = tracefile.get_df("top1_s.tr")
df1m = tracefile.get_df("top1_m.tr")
df1l = tracefile.get_df("top1_l.tr")

```

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.array([5, 10, 15])

y1 = np.array([tracefile.packet_delivery_ratio(df1s), tracefile.packet_delivery_ratio(df1m), tracefile.packet_delivery_ratio(df1l)])

y2 = np.array([tracefile.packet_delivery_ratio(df2s), tracefile.packet_delivery_ratio(df2m), tracefile.packet_delivery_ratio(df2l)])


fig, ax = plt.subplots()

ax.plot(x, y1, label='Distance Vector', marker="D")

ax.plot(x, y2, label='Link State', marker="D")

ax.legend()
plt.show()
```
